## Imports

Mari memulai dengan melakukan import pada modul Tensorflow dan kita berikan alias tf untuk mempermudah pemanggilan.

Kemudian mengimport library numpy, yang akan digunakan untuk menampung data.

Kerangka kerja dari Neural Network merupakan sebuah lapisan layer yang sequential yang menggunakan library Keras jadi mari mengimport library tersebut.


````python
import tensorflow as tf
import numpy as np
from tensorflow import keras

print(tf.__version__)
````

## Define and Compile the Neural Network

selanjutnya kita akan membuat arsitektur Neural Network sederhana. dimana memiliki hanya 1 layer (1 neuron, dan input shape yang bernilai 1)

````python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
````


Sekarang kita akan melakukan compile terhadap Neural Network yang sudah dibuat sebelumnya. dimana dalam proses ini kita memerlukan 2 fungsi yaitu loss dan optimizer.

Jika biasanya anda melihat banyak penerapan matematika dalam neural network, maka disinilah hal tersebut diterapkan. namun, dalam kasus ini hal tersebut telah dikemas dalam sebuah fungsi- lets explain


kita mengetahui bahwa di dalam fungsi , hubungan antara deret angka yaitu y = 2x-1

komputer juga belajar untuk mengetahui persamaan tersebut, pertama komputer akan menebak mungkin persamaannya y=10x+10. kemudian LOSS function akan mengukur tebakan persamaan dan menghitung seberapa bagus dan buruknya persamaan tersebut. 

kemudian akan menggunakan OPTIMIZER function untuk membuat tebakan lainnya. berdasarkan hasil dari LOSS function maka optimizer akan berusaha untuk meminimalkan nilai dari loss. kemudian akan membuat persamaam kembali seperti y=5x+5, meskipun masih buruk namun sudah mendekati ke persamaan yang benar (contohnya : ketika nilaii loss sangat rendah)


proses diatas akan terus berulang sebanyak EPOCH yang didefinisikan. Disini kita menggunakan "MEAN SQUARED ERROR" untuk LOSS dan "STOCHASTIC GRADIENT DESCENT" untuk optimizer. untuk sekarang jangan berfikit terlalu jauh tentang bagaimana perhitungan didalamnya, tetapi lihatlah bahwa hal ini bekerja !

Seiring waktu, Anda akan mempelajari fungsi loss dan optimizer yang berbeda dan sesuai untuk skenario yang berbeda. 


````python
model.compile(optimizer='sgd', loss='mean_squared_error')
````

## Providing the Data

selanjutnya, kita akan menggunakan beberapa data. didalam kasus ini kita menggunakan 6 data x dan 6 data y. dimana sebenarnya kita bisa melihat hubungan diantara x dan y, yaitu y=2x-1. (ketika x=-1 maka y=-3, dan seterusnya)

Python menyediakan banyak library yang mampu menghandle data salah satunya yaitu Numpy yang menyediakan struktur data bertipe array.

````python
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
````

# Training the Neural Network


selama proses training Neural Network, akan "belajar" mengetahui hubungan diantara Xs dan Ys ketika **model.fit** dipanggil.

````python
model.fit(xs, ys, epochs=500)
````



Ok, sekarang model sudah berlatih untuk mengetahui hubungan diantara Xs dan Ys. selanjutnya, kita akan menggunakan **mode.predict** untuk mencari nilai Y dari nilai X yang akan diprediksi. 

````python
print(model.predict([10.0]))
````

Anda mungkin berpikir 19, bukan? Tapi itu akhirnya menjadi sedikit di bawahnya. Menurut Anda mengapa demikian?

Ingat bahwa jaringan saraf berurusan dengan probabilitas, jadi dengan data training yang kita masukkan ke NN, dihitung bahwa ada kemungkinan yang sangat tinggi bahwa hubungan antara X dan Y adalah Y=2X-1, tetapi hanya dengan 6 titik data kita bisa tidak tahu pasti. Akibatnya, hasil untuk 10 sangat mendekati 19, tetapi belum tentu 19.

Saat Anda bekerja dengan jaringan saraf, Anda akan melihat pola ini berulang. Anda hampir selalu berurusan dengan probabilitas, bukan kepastian, dan akan melakukan sedikit pengkodean untuk mencari tahu apa hasilnya didasarkan pada probabilitas, terutama dalam hal klasifikasi.
