```python
# Constructor Example

class Apple:
  def __init__(self, color, flavor):
    self.color = color
    self.flavor = flavor

jonagold = Apple("Red", "delicious and very sweet")
```

```python
# Special Method __str__

class Apple:
  def __init__(self, color, flavor):
    self.color = color
    self.flavor = flavor
  def __str__(self): #The __str__ method in Python represents the class objects as a string
    return "This Apple is {} and its flavor is {}".format(self.color, self.flavor)
    
```

```python
jonagold = Apple("Red", "very sweet !")
print(jonagold)
```

# Documenting Functions, Classes, and Methods

```python
class Apple:
  def __init__(self, color, flavor):
    self.color = color
    self.flavor = flavor
  def __str__(self): #The __str__ method in Python represents the class objects as a string
    return "This Apple is {} and its flavor is {}".format(self.color, self.flavor)
```

```python
help(Apple)
```

```python
# tambahkan dokumentasi agar mudah dipahami menggunakan docstring
def to_second(hours, minutes, seconds):
  """Return the amount of seconds in the given hours, minutes, and seconds"""
  return hours*3600+minutes*60+seconds
```

```python
help(to_second)
```

# Code Reuse



---
Inheritance


```python
# Example Inheritance 1

class Fruits:
  def __init__(self, color, flavor):
    self.flavor = flavor
    self.color = color
```

```python
class Apple(Fruits):
  pass
```

```python
class Banana(Fruits):
  pass
```

```python
manalagi = Apple("Red", "sweet")
ambon = Banana("Yellow", "sweet")
```

```python
print(manalagi.color)
```

```python
print(ambon.color)
```

```python
# Example inheritance 2
class Animal:
  sound = ""
  def __init__(self, name):
    self.name = name
  def speak(self):
    print("{sound}!, I'm {name}! {sound}".format(sound=self.sound, name=self.name))
```

```python
class Kitty(Animal):
  sound = "Miyaw!"
```

```python
anabul = Kitty("anabul")
anabul.speak()
```

```python
class Cow(Animal):
  sound = "Mooooooooo"
```

```python
sapi = Cow("sapi")
sapi.speak()
```
