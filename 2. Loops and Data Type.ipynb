```python
# INFINITIVE LOOPS AND HOW TO BREAK THEM ? 
while x % 0 == 0 :
  x = x / 2
```

```python
# example infinitive loops :
x = 0
while x % 2 == 0 :
  x = x / 2
  print("hello")
```

```python
# handle this problem 

x = 4 
if x != 0:
  while x % 2 == 0 :
    x = x / 2
    print("hello") 
    
```

```python
#  handle this problem using logical operator

x = 2
while x != 0 and x % 2 == 0 :
  x = x / 2
  print("hello")
```

# For Loops

```python
my_friends = ['jojo', 'blacky', 'jeje', 'diablo']

for friend in my_friends:
  print("Hi " + friend)
```

```python
# another example for loops can iterate over a list of numbers to calculate the total sum and average
values = [1, 2, 3, 4, 5, 6, 7]

sum = 0
length = 0
for value in values : 
  sum += value
  length += 1

print("Total sum: " + str(sum) + " Total Average :" + str(sum/length))
```

```python
# FOR LOOPS USING "IN RANGE"

def convert_to_celcius(x):
  return (x-32)*5/9

for x in range(0, 101, 10):
  print(x, convert_to_celcius(x))
```

```python
# NESTED FOR LOOP (LOOP DI DALAM LOOP)

teams = ['persibaya', 'arema', 'bali united', 'borneo', 'persela']
for home_team in teams: 
  for away_team in teams:
    if home_team != away_team: 
      print(home_team + " VS " + away_team)
```

```python
# another example

for left in range(7):
  for right in range(left, 7):
    print("[" + str(left) + "|" + str(right) + "]", end=" ")
  print(" ")
```

```python
# COMMON ERROR WHEN USING FOR LOOP

# 1. DONT FORGET TO USE RANGE KEYWORD

for x in range(5):
  print(x)
```

```python
# 2. when you want to iterate only one element using one list

for x in [25]:
  print(x)
```

```python
# 3. more example

def greeting(friends):
  for friend in friends:
    print("Hi " + friend)

greeting("Bambang")
```

# Recursion

```python
def factorial(n):    
  print("Factorial called with " + str(n))
  if n<2:
    print("returning 1")
    return 1
  result = n * factorial(n-1)
  print("Returning " + str(result) + " for factorial of " + str(n))
  return result

factorial(5)
```

```python
# Real World Problem

def count_letter(text):
  result = {}
  for letter in text:
    if letter not in result:
      result[letter] = 0
    result[letter] += 1
  return result


count_letter("now we are learning about dictionary")
```
