# Function

```python
def greeting(name):
  print("Hello Good Morning, "+ name)

greeting("Lorenzo")
greeting("Diana")
```

```python
# Using Return one value in Function
def area_triangle(base, height):
  return (base*height)/2 

area_a = area_triangle(5,6)
area_b = area_triangle(3,5)
sum = area_a + area_b
print("The sum of both area is = " + str(sum))
```

```python
# return more than one value
def convert_second(seconds) :
  hours = seconds // 3600
  minutes = (seconds - hours * 3600) // 60
  remaining_seconds = seconds - hours * 3600 - minutes * 60
  return hours, minutes, remaining_seconds

convert_second(5000)
```

```python
hours, minutes, remaining_seconds = convert_second(5000)
```

```python
print(hours, minutes, remaining_seconds)
```

```python
print(hours)
```

```python
print(minutes)
```

```python
print(remaining_seconds)
```

```python
# PRINCIPLE OF CODE REUSE

name = "diana"
number = len(name) * 5
print("Hello " + name + " your lucky number is " + str(number))

name = "lorenzo"
number = len(name) * 5
print("Hello " + name + " your lucky number is " + str(number))
```

```python
def lucky_number(name):
  number = len(name) * 5
  print("Hello " + name + " your lucky number is " + str(number))

lucky_number("diana")
lucky_number("lorenzo")
```

```python
#CODE STYLE IS IMPORTANT

# before
def calculator(d):
  a = 3.14
  b = a * (d ** 2) 
  print(b)

calculator(5)
```

```python
# after
def calculator(radius):
  pi = 3.14 
  area_circle = pi * (radius ** 2)
  print("The area circle is = " + str(area_circle))

calculator(5)
```

```python
# ELSE STATEMENTS 1

def hint_username(username):
  if username < 3 :
    print("Invalid username. Must be at least 3 characters long")
  else:
    print("Valid username!")
```
